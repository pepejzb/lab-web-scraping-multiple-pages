2. Find the number of titles that have changed in the United States Code since its last release point: url = 'http://uscode.house.gov/download/download.shtml'

```python
# Importación de librerías
from bs4 import BeautifulSoup
import requests

# URL del sitio a scrapear
url= "http://uscode.house.gov/download/download.shtml"

# Solicitud de consulta URL
response = requests.get(url)

# Si el resultado es 200, está ok.
response.status_code
```

```python
# Esto me da todo el HTML de la página.
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
soup.select ('#us\/usc\/t1')
```

```python
soup.find_all ('div', class_= 'usctitlechanged')
```
